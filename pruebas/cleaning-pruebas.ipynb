```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
```

```python
diamonds=pd.read_csv('./input/data.csv')
diamonds.head()
```

```python
diamonds.cut.unique()
```

```python
diamonds.clarity.unique()
```

```python
diamonds.color.unique()
```

```python
diamonds.isnull().any()
```

```python
diamonds=pd.get_dummies(diamonds)
```

```python
diamonds.head()
```

```python
diamonds.columns
```

```python
corrMatrix = diamonds.corr()
corrMatrix
```

```python
fig, ax = plt.subplots(figsize=(20,20)) 
heatmap=sns.heatmap(corrMatrix, annot=True)
heatmap
```

```python
#carat tiene alta correlación con x,y,z, que a su ver tienen alta correlación entre ellas,
#por lo que podemos eliminar las tres últimas.
#Carat y GT mucha correlación

diamonds.drop(columns=['x','z','y'],inplace=True)
```

```python
corrMatrix = diamonds.corr()
fig, ax = plt.subplots(figsize=(20,20)) 
heatmap=sns.heatmap(corrMatrix, annot=True)
heatmap
```
